```
sum=0

for x in range(1,101):
  sum +=x # sum=sum+x 0+1 0+2 0+3~+100
  
print(sum)
```

```
sum=0

for x in range(1,101,2):
  sum +=x # sum=sum+x  1+3+5+7+9+11~+99
  
print(sum)
```

```
total=1

for x in range(1,101,2):
  total *=x # 1x3x5x7x9x11x13x15~x99
  
print(total)
```

```
sum = 0 #定初始值為0
x=1 #從1開始

while x < 101: # x不得大於100
  sum +=x #sum=sum+x
  x = x+1 # x=變數,+1是從1開始
  # x += 1
  
print(sum)
```

```
sum = 0 #定初始值為0
x=1 #從1開始

while x < 101: # x不得大於100
  sum +=x #sum=sum+x
  x = x+2 # x=變數,x=1+2(3)~x=5+2(7)
  # x += 1
  
print(sum)
```

```
total = 1
x=1

while x < 101:
  total *=x #乘法=變數
  x = x+2     # x += 2 #1x3=3 ,1x5=5,1x7=7
  
print(total)
```
